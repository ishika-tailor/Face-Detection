```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_classifier=cv2.CascadeClassifier('Desktop/haarcascade_frontalface_default.xml')
```

```python
image=cv2.imread('Desktop/Unknown-9.jpg')
plt.imshow(image)
```

```python
img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
faces=face_classifier.detectMultiScale(img,1.3,5)
if faces is ():
    print('No faces detected')
```

```python
def detect_face(fix_img):
    face_rects=face_classifier.detectMultiScale(img)
    for(x,y,w,h) in face_rects:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),10)
        return img
```

```python
result=detect_face(img)
plt.imshow(result)
```
